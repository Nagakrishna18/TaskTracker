```python
tasktracker = open("tasks.txt", "a+")

task_status = input("HEY! what do you want to do now? (add/delete - del/updatestatus - update/removeall/checkstatus - status / list_tasks ) ")
i = 0
task = ""
if task_status.lower() == "add":
    print("You have chosen to add tasks.")
    tasktracker.seek(0)  # Move to the beginning of the file
    tasks_no = tasktracker.readlines()
    taskno = int(input("Enter number of tasks you want to add: "))
    tasktracker.seek(0)  # Move to the beginning of the file
    new_tasks = []
    for i in range(taskno): 
        task = input("Enter your task: ")
        taskcompletion = input("Enter task completion status (yes/no): ")
        if taskcompletion.lower() == "yes":
            new_tasks.append("task" + str(len(tasks_no) + i  ) + "-" + task + "               : completed\n")
        else:
            new_tasks.append("task" + str(len(tasks_no) + i   ) + "-" + task + "               : not completed\n")
    tasktracker.writelines(new_tasks)
elif task_status.lower() == "del":
    print("You have chosen to delete tasks.")
    taskno = int(input("Enter number of tasks you want to delete: "))
    tasks_to_delete = []
    for n in range(taskno):
        task_num = int(input("Enter the task number you want to delete: "))
        tasks_to_delete.append(task_num)
    tasktracker.seek(0)
    lines = tasktracker.readlines()
    for task_num in sorted(tasks_to_delete, reverse=True):
        if 0 < task_num <= len(lines):
            del lines[task_num]
        else:
            print("Task number out of range.")
    tasktracker.seek(0)
    tasktracker.truncate()
    tasktracker.writelines(lines)
elif task_status.lower() == "update":
    print("you have chosen to update task status.")
    taskno = int(input("Enter number of tasks you want to update: "))
    status_change = []
    for n in range (taskno):
        task_num = int(input("enter the task number you want to update: "))
        taskcompletion = input("Enter new task completion status (yes/no): ")
        status_change.append(task_num)
    tasktracker.seek(0)
    status = tasktracker.readlines()
    for task_num in sorted(status_change , reverse=True):
        if 0 < task_num <= len(status):
            if taskcompletion.lower() == "yes":
                  status[task_num ]= status[task_num ].replace("not completed", "completed")
            elif taskcompletion.lower() == "no":
                 status[task_num  ] = status[task_num ].replace("completed", "not completed")
                        
        else:
            print("Task number out of range.") 
        tasktracker.seek(0)
        tasktracker.truncate()
        tasktracker.writelines(status)
elif task_status.lower() == "removeall":
    print("You have chosen to remove all tasks.")
    tasktracker.seek(0)
    tasktracker.truncate()
elif task_status.lower() == "status":
    print("you have chosen to check task status.")
    task_no = int(input("Enter number of tasks you want to check: "))
    for task_num in range(task_no) :
        task_num = int(input("Enter the task number you want to check: "))
        tasktracker.seek(0)
        tasks = tasktracker.readlines()
        if 0 < task_num <= len(tasks):
           if "completed" in tasks[task_num]:
                print(f"Task {task_num} is completed.")
           elif "Not Completed" in tasks[task_num]:
                print(f"Task {task_num} is not completed.")
        else:
            print("Task number out of range.")
elif task_status.lower() == "list_tasks":
    taskstat = input("what tasks do you want to list?(all/completed/not completed) ")  
    if taskstat.lower() == "all":
        tasktracker.seek(0)
        tasks = tasktracker.readlines()
        print("Here are all your tasks:\n" +(tasks))  
    elif taskstat.lower() == "completed":
        tasktracker.seek(0)
        tasks = tasktracker.readlines()
        completed_tasks = [task.split(':')[0] for task in tasks if "completed" in task and 'not' not in task]
        print("Here are your completed tasks: " + str(completed_tasks)) 
    elif taskstat.lower() == "not completed":
        tasktracker.seek(0)
        tasks = tasktracker.readlines()
        notcompleted_tasks = [task.split(':')[0] for task in tasks if "not completed" in task]
        print("Here are your pending tasks:" + str( notcompleted_tasks))

    
    
        
               
    
    
tasktracker.close()

```
